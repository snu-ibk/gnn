#### relation 끼리 동일성 확인

```python
amazon_original_data.keys()
```

```python
net_upu_sp_matrix = amazon_original_data['net_upu']
net_uvu_sp_matrix = amazon_original_data['net_uvu']
net_usu_sp_matrix = amazon_original_data['net_usu']

homo_sp_matrix = amazon_original_data['homo']
```

```python
import numpy as np
from scipy.sparse import csc_matrix

# 두 희소 행렬이 동일한지 확인하는 함수
def sparse_matrices_equal(mat1, mat2):
    if mat1.shape != mat2.shape:
        return False
    if mat1.nnz != mat2.nnz:  # 저장된 요소의 개수가 같은지 확인
        return False
    # 실제 데이터, 인덱스, 인덱스 포인터 비교
    return (np.array_equal(mat1.data, mat2.data) and
            np.array_equal(mat1.indices, mat2.indices) and
            np.array_equal(mat1.indptr, mat2.indptr))
```

```python
print('upu & uvu ', sparse_matrices_equal(net_upu_sp_matrix, net_uvu_sp_matrix))
print('upu & usu ', sparse_matrices_equal(net_upu_sp_matrix, net_usu_sp_matrix))
print('uvu & usu ', sparse_matrices_equal(net_uvu_sp_matrix, net_usu_sp_matrix))

print('upu & homo ', sparse_matrices_equal(net_upu_sp_matrix, homo_sp_matrix))
print('uvu & homo ', sparse_matrices_equal(net_uvu_sp_matrix, homo_sp_matrix))
print('usu & homo ', sparse_matrices_equal(net_usu_sp_matrix, homo_sp_matrix))
```

```python
# 세 개의 희소 행렬의 union을 계산
union_matrix = net_upu_sp_matrix.copy()
union_matrix += net_uvu_sp_matrix
union_matrix += net_usu_sp_matrix

# 값이 2 이상인 요소를 1로 변환하여 union 연산의 결과를 얻음
union_matrix.data = np.where(union_matrix.data > 1, 1, union_matrix.data)
```
